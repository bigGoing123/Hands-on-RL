```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# 加载训练集
training_data = pd.read_csv('training_set.csv')

X=training_data.drop(columns=['Unnamed: 0','ID','Customer_ID','Name','SSN','Credit_Score','Payment_of_Min_Amount',
                              'Payment_Behaviour','Num_of_Loan','Num_Bank_Accounts'])
# 获取所有类别型数据的列名
categorical_columns = X.select_dtypes(include=['object']).columns.tolist()

# 创建 LabelEncoder 对象
label_encoder = LabelEncoder()

# 使用 apply 对所有类别型数据列同时进行标签编码
X[categorical_columns] = X[categorical_columns].apply(lambda col: label_encoder.fit_transform(col))
y=training_data['Credit_Score']
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=42)
forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)  # n_estimators是随机森林中树的数量
forest_classifier.fit(X_train, y_train)
# 在验证集上进行预测
val_predictions = forest_classifier.predict(X_val)
# 计算验证集的准确率
accuracy = accuracy_score(y_val, val_predictions)
print(f"Val_accuracy: {accuracy:.4f}")
test_data=pd.read_csv('test/test_set.csv')
XX=test_data.drop(columns=['Unnamed: 0','ID','Customer_ID','Name','SSN','Credit_Score','Payment_of_Min_Amount',
                              'Payment_Behaviour','Num_of_Loan','Num_Bank_Accounts'])
# 获取所有类别型数据的列名
categorical_columns = XX.select_dtypes(include=['object']).columns.tolist()

# 创建 LabelEncoder 对象
label_encoder1 = LabelEncoder()

# 使用 apply 对所有类别型数据列同时进行标签编码
XX[categorical_columns] = XX[categorical_columns].apply(lambda col: label_encoder1.fit_transform(col))
test_predictions = forest_classifier.predict(XX)
test_data['Credit_Score']=test_predictions
test_data.to_csv('test/test_set.csv', index=False) # 保存回 existing_file.xlsx 文件
```
